# Bank Nifty Heatmap and EDA

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import calendar
```

```python
df = pd.read_csv("../input/banking/banknifty-data (1).csv")
```

```python
df.head()
```

```python
df['datetime'] = pd.to_datetime(df['datetime'])
```

```python
df.dtypes
```

```python
values_needed = ['open', 'close']
temp_df = df.groupby([df['datetime'].dt.year,df['datetime'].dt.month])[values_needed].sum()
```

```python
temp_df=temp_df.reset_index(level=1)
temp_df=temp_df.rename(columns={'datetime':'months'})
temp_df=temp_df.reset_index()
temp_df=temp_df.rename(columns={'datetime':'years'})
```

## Calculating Month % Return
 formula = (Month Open - Month Close)*100/Month Open

```python
temp_df['MonthlyReturn'] = (temp_df['close']-temp_df['open'])*100/temp_df['open']
```

```python
columns = ['years','months','MonthlyReturn']
monthly_df = temp_df[columns]
```

```python
monthly_df
```

```python
monthly_df['months']
monthly_df["months"] = monthly_df.months.apply(lambda x: calendar.month_abbr[x])
```

```python
x = np.array([ str(calendar.month_abbr[x]) for x in range(1,13)])
x = x.astype('object')
```

```python
mh = monthly_df.pivot("years", "months", "MonthlyReturn")
mh=mh.loc[:, x]
plt.figure(figsize=(16,12))
sns.heatmap(mh, annot=True, cmap='RdYlGn')

_=plt.title("Monthly Returns (%)", fontdict = {'fontsize': 20})
```

```python
#noticing the great recession of 2008
plt.figure(figsize = (14,10))
sns.lineplot(data=monthly_df, x = "years", y = "MonthlyReturn", hue = "months", palette="hsv",legend='full')
_=plt.title("Trend by every year", fontdict = {'fontsize': 20})
```

```python
plt.figure(figsize = (14,10))
sns.lineplot(data=monthly_df, x = "months", y = "MonthlyReturn", hue = "years", palette="hsv",legend='full')
_=plt.title("Trend by every months", fontdict = {'fontsize': 20})
```

```python
#data of 2008 vs 2009
temp_df_2008 = monthly_df[monthly_df["years"] == 2008]
temp_df_2009 = monthly_df[monthly_df["years"] == 2009]
sns.lineplot(data=temp_df_2008, x = "months", y = "MonthlyReturn")
sns.lineplot(data=temp_df_2009, x = "months", y = "MonthlyReturn" )
```

```python
#data of 2008 vs 2009
temp_df_2018 = monthly_df[monthly_df["years"] == 2018]
temp_df_2019 = monthly_df[monthly_df["years"] == 2019]
sns.lineplot(data=temp_df_2018, x = "months", y = "MonthlyReturn")
sns.lineplot(data=temp_df_2019, x = "months", y = "MonthlyReturn" )
```

```python
#removing outlier 2008 as the world was going through recession
#removing outlier 2009 as this year was the major recovery year
monthly_df = monthly_df[monthly_df["years"]!=2008]
monthly_df = monthly_df[monthly_df["years"]!=2009]
```

```python
#avg return of every month
avgMR_df=pd.DataFrame()
for month in x:
    avgMR_df=avgMR_df.append({"months":month, "monthlyreturn": monthly_df[monthly_df["months" ]== month].sum()["MonthlyReturn"]},ignore_index=True)
```

```python
sns.lineplot(data=avgMR_df, x="months", y = "monthlyreturn")
```

We notice that on average duing the spring season and winter season the returns are healthy.
